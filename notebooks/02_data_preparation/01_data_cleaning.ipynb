# Data Cleaning

```python
import numpy as np
import pandas as pd

train_df = pd.read_csv("../../data/raw/train.csv")
test_df = pd.read_csv("../../data/raw/test.csv")
```

```python
trainset_df = train_df.dropna(subset=["Age","Annual Income","Marital Status","Number of Dependents","Health Score","Customer Feedback","Vehicle Age","Insurance Duration"])
trainset_df  = trainset_df.drop(columns=["id","Occupation","Previous Claims","Credit Score","Policy Start Date"])
trainset_df = trainset_df.reset_index(drop=True)

print(trainset_df.shape)
trainset_df.head(10)
```

```python
trainset_df.isna().sum()
```

```python
trainset_df.to_csv("../../data/processed/train_data_cleaned.csv", index=False)
```

- Se eliminaron los valores faltantes de las columnas "Age","Annual Income","Marital Status","Number of Dependents","Health Score","Customer Feedback","Vehicle Age","Insurance Duration".
- Se eliminaron las siguientes columnas debido a tener gran cantidad de datos faltantes "id","Occupation","Previous Claims","Credit Score"
- Se elimino la siguiente columna "Policy Start Date" en versiones posteriores se planea realizar un mejor an√°lisis.
