# EDA Detallado del Dataset de Seguros
En este Notebook se analiza como van cambiando los datos originales luego de pasar por los procesos de limpieza y transformaci√≥n.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("../../data/raw/train.csv")
```

```python
train_clean_df = pd.read_csv("../../data/processed/train_data_cleaned.csv")
```

```python
train_transform_df = pd.read_csv("../../data/processed/train_data_encoded.csv")
```

### Datos Originales

```python
train_df.hist(bins=50, figsize=(25,8),layout=(3,4))
plt.tight_layout()
plt.show()
```

### Datos Limpiados

```python
train_clean_df.hist(bins=50, figsize=(25,8),layout=(3,4))
plt.tight_layout()
plt.show()
```

### Datos Transformados

```python
train_transform_df.hist(bins=50, figsize=(25,12),layout=(4,4))
plt.tight_layout()
plt.show()
```

- Se observa que "Premium Amount" ahora tiene un un histograma con una distribuci√≥n mas normal, lo que evitar√≠a el sesgo.

```python
sns.histplot(train_transform_df["Premium Amount"], kde=True)
```

## Visualizaci√≥n de las variables Categ√≥ricas
Histograma de las variables categ√≥ricas, en este caso se omite la "Policy Start Date"

```python
fig, axes = plt.subplots(2, 5, figsize=(15,7))
train_df["Gender"].value_counts().plot(kind="bar", ax=axes[0,0])
train_df["Marital Status"].value_counts().plot(kind="bar", ax=axes[0,1])
train_df["Education Level"].value_counts().plot(kind="bar", ax=axes[0,2])
train_df["Occupation"].value_counts().plot(kind="bar", ax=axes[0,3])
train_df["Location"].value_counts().plot(kind="bar", ax=axes[0,4])
train_df["Policy Type"].value_counts().plot(kind="bar", ax=axes[1,0])
train_df["Smoking Status"].value_counts().plot(kind="bar", ax=axes[1,1])
train_df["Exercise Frequency"].value_counts().plot(kind="bar", ax=axes[1,2])
train_df["Property Type"].value_counts().plot(kind="bar", ax=axes[1,3])
train_df["Customer Feedback"].value_counts().plot(kind="bar", ax=axes[1,4])
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15,7))
train_clean_df["Gender"].value_counts().plot(kind="bar", ax=axes[0,0])
train_clean_df["Marital Status"].value_counts().plot(kind="bar", ax=axes[0,1])
train_clean_df["Education Level"].value_counts().plot(kind="bar", ax=axes[0,2])
train_clean_df["Location"].value_counts().plot(kind="bar", ax=axes[0,4])
train_clean_df["Policy Type"].value_counts().plot(kind="bar", ax=axes[1,0])
train_clean_df["Smoking Status"].value_counts().plot(kind="bar", ax=axes[1,1])
train_clean_df["Exercise Frequency"].value_counts().plot(kind="bar", ax=axes[1,2])
train_clean_df["Property Type"].value_counts().plot(kind="bar", ax=axes[1,3])
train_clean_df["Customer Feedback"].value_counts().plot(kind="bar", ax=axes[1,4])
plt.tight_layout()
plt.show()
```

```python
corr = train_df[["Age","Annual Income","Number of Dependents","Health Score","Previous Claims","Vehicle Age","Credit Score","Insurance Duration","Premium Amount"]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
